```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city=pd.read_csv(city_data_to_load)
ride=pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
city_ride=pd.merge(city, ride, on="city")
# Display the data table for preview
city_ride.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
city_ride_urban=city_ride.loc[city_ride["type"]=="Urban"]
city_ride_suburban=city_ride.loc[city_ride["type"]=="Suburban"]
city_ride_rural=city_ride.loc[city_ride["type"]=="Rural"]
number_rides_urban=city_ride_urban.groupby(["city"])["ride_id"].count()
average_fare_urban=city_ride_urban.groupby(["city"])["fare"].mean()
driver_count_urban=city_ride_urban.groupby(["city"])["driver_count"].mean()

number_rides_suburban=city_ride_suburban.groupby(["city"])["ride_id"].count()
average_fare_suburban=city_ride_suburban.groupby(["city"])["fare"].mean()
driver_count_suburban=city_ride_suburban.groupby(["city"])["driver_count"].mean()

number_rides_rural=city_ride_rural.groupby(["city"])["ride_id"].count()
average_fare_rural=city_ride_rural.groupby(["city"])["fare"].mean()
driver_count_rural=city_ride_rural.groupby(["city"])["driver_count"].mean()
# Build the scatter plots for each city types
urban_plot=plt.scatter(number_rides_urban, average_fare_urban, s=driver_count_urban, c="lightcoral",edgecolors="black",alpha=0.8,linewidths=0.5)
suburban_plot=plt.scatter(number_rides_suburban, average_fare_suburban, s=driver_count_suburban, c="lightskyblue",edgecolors="black",alpha=0.8,linewidths=0.5)
rural_plot=plt.scatter(number_rides_rural, average_fare_rural, s=driver_count_rural, c="gold",edgecolors="black",alpha=0.8,linewidths=0.5)
# Incorporate the other graph properties
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")
plt.grid(b=None, which='major', axis='both')
# Create a legend
plt.legend((urban_plot, suburban_plot, rural_plot), ('Urban', 'Suburban', 'Rural'),title="City Types")
# Incorporate a text label regarding circle size
plt.text(x=1,y=1,s="Note:Circle size correlates with driver count per city")
# Save Figure
plt.savefig("Pyber.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
urban_total_fare=city_ride_urban["fare"].sum()
suburban_total_fare=city_ride_suburban["fare"].sum()
rural_total_fare=city_ride_rural["fare"].sum()
color=["lightcoral","lightskyblue","gold"]
label=["Urban","Suburban","Rural"]
explode=[0.1,0,0]

# Build Pie Chart
plt.pie([urban_total_fare,suburban_total_fare,rural_total_fare],explode=explode,labels=label,colors=color,autopct="%1.1f%%",shadow=True, startangle=140)
plt.title("% of Total Fares by City Type")
# Save Figure
plt.savefig("totalfarepie.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
urban_total_rides=city_ride_urban["ride_id"].count()
suburban_total_rides=city_ride_suburban["ride_id"].count()
rural_total_rides=city_ride_rural["ride_id"].count()
color=["lightcoral","lightskyblue","gold"]
label=["Urban","Suburban","Rural"]
explode=[0.1,0,0]

# Build Pie Chart
plt.pie([urban_total_rides,suburban_total_rides,rural_total_rides],explode=explode,labels=label,colors=color,autopct="%1.1f%%",shadow=True, startangle=140) 
plt.title("% of Total Rides by City Type")
# Save Figure
plt.savefig("totalridespie.png")

```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
urban_total_drivers=city_ride_urban["driver_count"].sum()
suburban_total_drivers=city_ride_suburban["driver_count"].sum()
rural_total_drivers=city_ride_rural["driver_count"].sum()
color=["lightcoral","lightskyblue","gold"]
label=["Urban","Suburban","Rural"]
explode=[0.1,0,0]

# Build Pie Chart
plt.pie([urban_total_drivers,suburban_total_drivers,rural_total_drivers],explode=explode,labels=label,colors=color,autopct="%1.1f%%",shadow=True, startangle=140) 
plt.title("% of Total Drivers by City Type")
# Save Figure
plt.savefig("totaldriverspie.png")
```

```python
# Show Figure
plt.show()
```
